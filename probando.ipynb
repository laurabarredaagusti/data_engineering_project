```python
import sqlite3
import pandas as pd
```

```python
#1) Llamar a la base de datos:
connection = sqlite3.connect("database.sqlite")
crsr = connection.cursor()
```

```python

#2) Definir función:
def sql_query(query):

    # Ejecuta la query
    crsr.execute(query)

    # Almacena los datos de la query 
    ans = crsr.fetchall()

    # Obtenemos los nombres de las columnas de la tabla
    names = [description[0] for description in crsr.description]

    return pd.DataFrame(ans,columns=names)
```

```python
#3) Definir query para coger todos los datos:
query = '''
        SELECT * 
        FROM Match
        '''
```

```python
#4) Llama a la función con la query y crea dataframe
df = sql_query(query)
df.head()
```
